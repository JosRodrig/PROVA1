```python
while True:
    idade = int(input("Digite sua idade: "))
    valor_total = float(input("Digite valor disponivel em carteira R$: "))
    ingresso = 30.00

    print (f"Valor do ingresso: R$ {ingresso:.2f}")

    if idade < 16:
        print("Entrada permitida apenas com acompanhante!")

    if valor_total >= ingresso:
        print("Entrada liberada!")
    else:
        print("Saldo insuficiente para o ingresso.")

    repeticao = input("Verificar outro cliente (s/n): ")
    if repeticao == "n":
        print("Programa encerrado!")
        break
    print()

```
